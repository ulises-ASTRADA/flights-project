```python
import requests
```

```python
import pandas as pd
```

```python
#Here we define the constants to connect to the API
ACCESS_KEY = '13824a221d87297fe2e6bc3c8ee20a94'
BASE_URL = 'https://api.aviationstack.com/v1/'
```

```python
#Request to a filtered API with the necessary parameters and storing the results as a json
api_result = requests.get(f'{BASE_URL}flights?access_key={ACCESS_KEY}&limit=100&flight_status=active')

flights = api_result.json()
```

```python
#Here we access the columns we are interested in
flights_data = flights['data']
print(flights_data)
```

```python
def data_to_pandas(flights_data):
    '''This function takes a json as an argument, change some characters and convert the 
    it to a pandas dataframe then download the result'''
    collected_data = []

    for flight in flights_data:
        flight_info = {
            'flight_date': flight['flight_date'],
            'flight_status': flight['flight_status'],
            'departure_airport': flight['departure']['airport'],
            'departure_timezone': flight['departure']['timezone'],
            'arrival_airport': flight['arrival']['airport'],
            'arrival_timezone' : flight['arrival']['timezone'],
            'arrival_terminal' : flight['arrival']['terminal'],
            'airline_name': flight['airline']['name'],
            'flight_number': flight['flight']['number']
        }
    
        collected_data.append(flight_info)
        collected_data_to_df = pd.DataFrame(collected_data)
        collected_data_to_df = collected_data_to_df.replace('/', '-', regex=True)
        collected_data_to_df.to_csv('/Users/naranax/Desktop/projects/flight-project/new_csv.csv', index=False)
        
        number_of_rows = len(collected_data_to_df)
    print(f'Data downloaded containing {number_of_rows} rows')
data_to_pandas(flights_data)
```
